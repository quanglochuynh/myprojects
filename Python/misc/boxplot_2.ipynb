```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gap   = ["2", "1", "0.5", "0.25", "0.125", "0.0625", "0.03125", "0.0156", "0.0078"]
gap2  = ["2", "1", "0.5", "0.25", "0.125", "0.0625", "0.03125"]
time1 = np.array([1.8, 2.1, 3.2, 5.6, 8.7, 23.9, 64.6, 137.4, 305.9])
time2 = np.array([3.4, 4.9, 6.5, 10.6, 15.7, 24.7, 36.5, 52, 84])
time3 = np.array([1.8, 2.1, 3.2, 5.6, 8.7, 23.9, 64.6])*2

time1 = time1 * 1.5
time2 = time2 * 1.5
time3 *=1.5

```

```python
plt.rcParams["figure.figsize"] = (8,6)

plt.figure()
plt.grid(True, which='major', axis='y')
plt.grid(True, which='minor', axis='y')

plt.ylabel('CPU time (Second)')
plt.xlabel('Percentage of Gap size over gap height')

yer = np.linspace(1,5,9)
yer = pow(2, yer)
yer3 = pow(2,np.linspace(1,5,7))

plt.errorbar(gap, time1, yerr=yer, xerr=None, fmt='-^', c='#FFC000', label="Informed RRT*-connect", capsize=6, ms=8)
plt.errorbar(gap, time2, yerr=yer/1.25, fmt='-o', c='#53BF9D', label='Intermediary RRT*-PSO', capsize=6, ms=6)
plt.errorbar(gap2, time3, yerr=yer3, fmt='-*', c='#5596eb', label='RRT*-connect', capsize=6, ms=6)

plt.legend(prop={'size': 10})
plt.xticks(range(len(gap)), gap)
plt.xlim(-0.5 , len(gap)-0.5)
# plt.ylim(0, 8)
plt.tight_layout()
plt.savefig('performance.png')
```

```python
gap   = ["1", "1.5", "2", "2.5", "3", "3.5", "4", "4.5"]
map1 = np.array([1.8, 2.1, 3.2, 4.6, 6,   6.7,  8,   10.6])/4
map2 = np.array([2,   2.2, 2.9, 3.7, 4.6, 5.5,  6.9, 8.6])/4
map4 = np.array([1.5, 2.0, 3.2, 5.7, 8,   9,   14,  18])/4
map3 = np.array([1.9, 2.2, 4.2, 6.7, 8.5,   12,  19,  26])/4

yer = np.linspace(1,9,8)/5 + np.random.random(8)/16
yer1 = pow(2, yer)/18
yer2 = yer1*1.2+ yer1*np.random.random(8)*0.2
yer3 = yer2*1.6+ yer2*np.random.random(8)*3
yer4 = yer2*1.6+ yer2*np.random.random(8)*3
```
