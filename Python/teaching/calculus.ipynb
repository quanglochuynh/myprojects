```python
# zero degree derivative
def func(x):
    # return 10 * np.power(np.e,-0.1*x) *np.sin(x) 
    # return np.sin(x)
    return np.power(2,x)
```

```python
# first degree derivative
def derive(x, fx):
    dx = 0.000001
    return (fx(x+dx)-fx(x))/dx
```
