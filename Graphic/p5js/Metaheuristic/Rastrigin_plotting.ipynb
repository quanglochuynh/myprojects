```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
```

```python
X = np.linspace(-400, 400, 100)     
Y = np.linspace(-400, 400, 100)     
X, Y = np.meshgrid(X, Y) 

Z = ((0.05*X)**2 - 100 * np.sin(0.02*np.pi * X)) + ((0.05*Y)**2 - 100 * np.cos(0.02*np.pi * Y))
# Z = np.sin(0.025*X) + np.sin(0.025*Y)
```

```python
plt.imshow(Z)
plt.show()
```

```python
fig = plt.figure() 
ax = fig.gca(projection='3d') 
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
  cmap=cm.nipy_spectral, linewidth=0.08,
  antialiased=True)    
# plt.savefig('rastrigin_graph.png')
plt.show()
```
